```python
# Importing initial libraries
import pandas as pd
```

```python
title_basics = pd.read_csv('Data/title_basics.csv.gz')
title_basics.info()
title_basics.head()
```

```python
## create a col with a list of genres
title_basics['genres_split'] = title_basics['genres'].str.split(',')
title_basics
```

```python
# Exploding the series
exploded_genres = title_basics.explode('genres_split')
exploded_genres
```

```python
# Taking only the unique entries after exploding
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
# saving just just genres and tconst as new df
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_basics.drop(columns='genres_split')
```

```python
# Manually making Dataframe
genre_lookup = pd.DataFrame({'Genre Name': genre_map.keys(), 'Genre_ID': genre_map.values()})
genre_lookup.head()
```

## Saving MYSQL tables

```python
from urllib.parse import quote_plus as urlquote
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
from sqlalchemy_utils import database_exists

## Change username and password to match your personal MySQL Server settings
username = 'root' # default username for MySQL db is root
password = 'Houriais@123' # whatever password you chose during MySQL installation.

connection = f'mysql+pymysql://{username}:{urlquote(password)}@localhost/sakila'
engine = create_engine(connection)
```

```python
## get max string length
basics_max_str_len = title_basics['genres'].fillna('').map(len).max()
```

```python
## Importing library
from sqlalchemy.types import *
## Calculate max string lengths for object columns
basics_key_len = title_basics['tconst'].fillna('').map(len).max()
basics_title_len = title_basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
basics_schema = {
    "tconst": String(basics_key_len+1), 
    "primaryTitle": Text(basics_title_len+1),
    'startYear':Float(),
    'runtimeMinutes':Integer(),
    "genres": Text(basics_max_str_len+1)}

```

```python
# Save to sql with dtype and index=False
title_basics.to_sql('title_basics', engine, dtype = basics_schema, if_exists='replace', index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

```python
## get max string length
ratings_max_str_len = title_ratings['tconst'].fillna('').map(len).max()

## Calculate max string lengths for object columns
ratings_key_len = title_ratings['tconst'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
ratings_schema = {
    "tconst": String(ratings_key_len+1), 
    'averageRating': Float(),
    'numVotes': Integer()}
```

```python
# Save to sql with dtype and index=False
title_ratings.to_sql('title_ratings', engine, dtype = ratings_schema, if_exists='replace', index=False)
```

```python
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
```

```python
## get max string length
tmdb_max_str_len = tmdb_results['imdb_id'].fillna('').map(len).max()

## Calculate max string lengths for object columns
tmdb_key_len = tmdb_results['imdb_id'].fillna('').map(len).max()
tmdb_title_len = tmdb_results['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
tmdb_schema = {
    "imdb_id": String(tmdb_key_len+1),
    "certification":Text(tmdb_title_len+1),
    'budget':Float(),
    'revenue':Float()}
```

```python
# Save to sql with dtype and index=False
tmdb_results.to_sql('tmdb_results', engine, dtype = tmdb_schema, if_exists='replace', index=False)
```
