```python
# Open saved file and preview again
basics = pd.read_csv("Data/title_basics.csv.gz", low_memory = False)
basics.head()
```

```python
#replacing the method they used to show NAs to an appropriate way for Python
basics.replace({'\\N':np.nan},inplace=True)

#checking how many NAs there are after
basics.isna().sum()
```

```python
#dropping NULL movies from runtimeMinutes and genres
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
basics.isna().sum()
```

```python
#selecting data based on titleType and startYear
basics = basics.loc[(basics['titleType'] =='movie')]
basics.head()
```

```python
#converting startYear to int
basics['startYear'] = basics['startYear'].fillna(0).astype(int)

basics = basics.loc[(basics['startYear'] >=2000) & (basics['startYear'] <=2020)]

basics
```

```python
basics.head()
```

```python
#eliminate movies that include "Documentary" in genre

# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
basics
```

### Filtering akas data set

```python
#checking out the structure of the akas data and how many NA are in currently
akas.info()
akas.isna().sum()
```

```python
akas.head()
```

```python
#replacing the method they used to show NAs to an appropriate way for Python
akas.replace({'\\N':np.nan},inplace=True)

#checking how many NAs there are after
akas.isna().sum()
```

```python
#keeping only US movies
akas = akas.loc[(akas['region'] =='US')]
akas
```

### Filtering basics data based on akas data

```python
#keeping movies from basics that are only in US filter from akas
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
#filtering basics with the new keepers filter
basics = basics[keepers]
basics
```

### Cleaning ratings data

```python
#checking out the structure of the ratings data and how many NA are in currently
ratings.info()
ratings.isna().sum()
```

```python
ratings.head()
```

```python
#replacing the method they used to show NAs to an appropriate way for Python
ratings.replace({'\\N':np.nan},inplace=True)

#checking how many NAs there are after
ratings.isna().sum()
```

```python
#keeping movies from ratings that are only in US filter from akas
keepers2 =ratings['tconst'].isin(akas['titleId'])
keepers2
```

```python
#filtering basics with the new keepers filter
ratings = ratings[keepers2]
ratings
```

```python
#saving copy of new filtered data into data folder
basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
akas.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
ratings.to_csv("Data/title_ratings.csv.gz",compression='gzip',index=False)

basics
```
